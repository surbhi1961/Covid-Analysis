```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
#surbhi
```

```python
df = pd.read_csv("C:/Users/ranes/OneDrive/Desktop/Covid-19-analysis/Covid cases in India.csv")
print(df.head())
```

```python
df.drop(['S. No.'], axis= 1, inplace=True)
print(df.head())
```

```python
df['Total Cases'] = df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
print(df.head())
```

```python
total_cases_overall = df['Total Cases'].sum()
print('Total number cases in India are',total_cases_overall)
```

```python
df['Active Cases']= df['Total Cases']-(df['Death']+df['Cured'])
print(df.head())
```

```python
Total_Active_Cases = df.groupby('Name of State / UT')['Total Cases'].sum().sort_values(ascending=True).to_frame()
Total_Active_Cases.style.background_gradient(cmap='blue')
print(Total_Active_Cases)
```

```python
df.plot(kind='bar', x='Name of State / UT', y='Total Cases')
plt.show()
```
